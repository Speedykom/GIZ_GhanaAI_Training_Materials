# Part 2: The Critical Bug We Will Fix

## The Problem

The crawler will find 0 links from <https://codetrainafrica.com/>
despite the HTML containing navigation links. After debugging, we will
discover:

**Root Cause:** The crawler will remove `<nav>` HTML elements to clean
up text content, but ALL the internal navigation links will be inside
those `<nav>` elements.

``` python
# BROKEN CODE - removed nav before extracting links
for script in soup(["script", "style", "nav", "footer"]):
    script.decompose()  # This removed the links!

# Then tried to find links
links = soup.find_all("a", href=True)  # Found 0 links
```

**The Fix:** Extract links FIRST, then clean up text.

``` python
# FIXED CODE - extract links before removing nav
# Step 1: Extract links
for link in soup.find_all("a", href=True):
    # ... process links ...

# Step 2: NOW remove script/style for text cleaning
for script in soup(["script", "style"]):
    script.decompose()
```

## Impact

-   Before fix: Will crawl 1 page, will find 0 links, will give generic
    responses
-   After fix: Will crawl multiple pages, will discover
    projects/blog/resume, will give specific answers


## CodeTrain Version (Python)

    git+https://github.com/Speedykom/CodeTrain.git
    fastapi==0.129.0
    uvicorn==0.41.0
    requests==2.32.5
    beautifulsoup4==2.8.3
    python-dotenv==1.2.1

### External Services Required

-   **OpenRouter API** (https://openrouter.ai/)
    -   Free tier available (Llama 3.1 70B)
    -   Or paid: ~\$0.01-0.05 per query
    -   API key required in `.env` file

# Part 4: What Worked vs. What Would Have Been Better

## Approach 1: Simple Static Chatbot (What We Started With)

**Implementation:**

-   Hardcoded responses in JavaScript
-   Keyword matching for queries
-   No backend required
-   Single HTML file

**Pros:**

-   ✅ Instant responses
-   ✅ Works offline
-   ✅ Zero cost
-   ✅ Simple to understand

**Cons:**

-   ❌ Static knowledge (stale quickly)
-   ❌ Limited to pre-programmed answers
-   ❌ Cannot explore new content
-   ❌ No understanding of context

**Best for:** Small FAQ bots, prototypes, budget-constrained projects

## Approach 2: Smart Scraper with API (The Middle Ground)

**Implementation:**

-   Node.js Express backend
-   Scrapes predefined URLs on each query
-   Keyword-based search across content
-   Returns best-matching sections

**Pros:**

-   ✅ Always fresh content
-   ✅ No AI costs (no LLM)
-   ✅ Fast enough for small sites
-   ✅ Can find relevant content

**Cons:**

-   ❌ No content synthesis (returns raw text)
-   ❌ Cannot decide what’s relevant
-   ❌ Limited to keyword matching
-   ❌ Cannot explore beyond predefined URLs

**Best for:** Medium sites (10-50 pages), cost-sensitive projects,
simple search

## Approach 3: CodeTrain Agent (What We Ended With)

**Implementation:**

-   Python FastAPI backend
-   Agent loop: Crawl → Decide → Answer/Explore
-   LLM makes intelligent decisions
-   Dynamic URL discovery
-   Content synthesis

**Architecture:**

    User Query → CrawlJob → DecideJob → 
      ├─ "Explore" → Find URLs → CrawlJob (loop)
      └─ "Answer" → AnswerJob → Synthesize Response

**Pros:**

-   ✅ Intelligent decision making
-   ✅ Dynamic exploration
-   ✅ Natural language synthesis
-   ✅ Discovers entire website
-   ✅ Context-aware responses
-   ✅ Scales to large sites

**Cons:**

-   ❌ Requires API key (costs money)
-   ❌ Slower (10-20s per query)
-   ❌ More complex architecture
-   ❌ Requires server infrastructure

**Best for:** Production chatbots, large websites, complex questions,
when quality matters

## The “Once and For All” Solution

**What we would build if starting fresh:**

### Hybrid Architecture

    User Query → Check Cache → 
      ├─ Cache Hit → Return cached answer
      └─ Cache Miss → 
          ├─ Vector Search (embeddings) → Find relevant chunks
          ├─ LLM Synthesis → Generate answer
          └─ Cache Answer → Store for 24 hours

**Key Components:**

1.  **Vector Database** (Pinecone, Weaviate, or Chroma)
    -   Pre-compute embeddings of website content
    -   Semantic search instead of keyword matching
    -   Fast retrieval (milliseconds)
2.  **Scheduled Crawler**
    -   Crawl website every 6-24 hours
    -   Update vector database automatically
    -   Keep content fresh without per-query crawling
3.  **Redis Cache**
    -   Cache LLM responses for common questions
    -   Reduce API costs by 80-90%
    -   Faster responses for repeat queries
4.  **Streaming Responses**
    -   Show LLM output as it’s generated
    -   Better user experience
    -   Feels more responsive
5.  **Conversation Memory**
    -   Track conversation history
    -   Context-aware follow-up questions
    -   More natural interactions

**Technology Stack:**

    Frontend: React or Vue.js
    Backend: FastAPI (Python)
    Vector DB: Pinecone or Chroma
    Cache: Redis
    LLM: OpenAI GPT-4 or Claude via OpenRouter
    Embeddings: OpenAI text-embedding-3-small
    Crawler: Playwright (for JS-heavy sites)
    Hosting: Docker + Kubernetes

**Cost Optimization:**

-   Vector search: ~\$0.001 per query
-   LLM: ~\$0.01 per query (with caching: ~\$0.002)
-   Total: ~\$0.003 per query vs \$0.05 without optimization
-   **10x cheaper** than naive implementation

**Best for:** Production at scale, enterprise, when cost and speed
matter

# Part 5: Complete File Structure

## Final Project Structure

    your-project-directory/
    ├── cursor-course-welcome.html      # Main website with integrated chatbot
    ├── server.js                        # Node.js backend (original)
    ├── package.json                     # Node dependencies
    ├── package-lock.json
    ├── README.md                        # Documentation
    ├── CODETRAIN_UPGRADE_SUMMARY.md    # Comparison document
    ├── COMPARISON.md                    # Architecture comparison
    ├── codetrain-chatbot/              # CodeTrain implementation
    │   ├── server.py                    # FastAPI backend
    │   ├── main.py                      # CLI entry point
    │   ├── jobs.py                      # CodeTrain Jobs
    │   ├── hustle.py                    # Workflow orchestrator
    │   ├── demo.py                      # CLI demo
    │   ├── requirements.txt             # Python dependencies
    │   ├── .env                         # API keys (gitignored)
    │   ├── .env.example                 # Example environment
    │   ├── .gitignore
    │   ├── README.md                    # Project docs
    │   ├── venv/                        # Virtual environment
    │   ├── utils/
    │   │   ├── web_crawler.py          # Fixed web scraper
    │   │   └── call_llm.py             # LLM integration
    │   ├── static/
    │   │   ├── index.html              # Standalone chat UI
    │   │   └── chatbot.js              # Frontend logic
    │   └── docs/
    │       └── design.md               # Architecture docs
    └── test_chatbot.html               # Test page

# Part 6: Access URLs and Testing

## Running the Complete System

### Terminal 1: Start CodeTrain API

``` bash
cd your-project-directory/codetrain-chatbot
source venv/bin/activate
./venv/bin/python server.py
# Runs on http://localhost:8000
```

# Part 7: Key Lessons Learned

## Technical Lessons

1.  **Order matters in web scraping** - Extract data before modifying
    HTML structure
2.  **CORS is tricky** - Opening files directly (`file://`) blocks
    localhost API calls
3.  **Always use a web server** - Even for “simple” HTML files
4.  **Clear Python cache** - `__pycache__` files hide code changes
5.  **Test incrementally** - Debug crawler separately from full system

## Architecture Lessons

1.  **Start simple** - Static chatbot proved the concept before adding
    complexity
2.  **Agent pattern is powerful** - Letting AI decide what to explore
    beats hardcoded logic
3.  **Cost vs. quality tradeoff** - LLM calls are expensive but provide
    better answers
4.  **Caching is essential** - Would save 80-90% of API costs in
    production
5.  **Vector search \> keyword search** - Semantic understanding beats
    exact matching

## Prompting Lessons

1.  **Be specific** - “Make it look awesome” worked because context was
    clear
2.  **Iterate** - Will start with OpenCode, will switch to CodeTrain
    Africa, will evolve to CodeTrain
3.  **Show examples** - Referencing CodeTrain repo provided clear
    direction
4.  **Debug together** - Walking through the crawler bug in real-time
    will fix it faster

``` yaml
services:

    name: sisengai-chatbot-api

    startCommand: cd backend && uvicorn server:app --host 0.0.0.0 --port $PORT

        source: /api/*
        destination: http://sisengai-chatbot-api.onrender.com/api/*
```

### Update Frontend API URL

``` javascript
// In frontend/index.html, change:
this.API_URL = 'http://localhost:8000/api';

// To:
this.API_URL = window.location.hostname === 'localhost' 
    ? 'http://localhost:8000/api' 
    : '/api';  // Same domain, will proxy to backend
```

### Configure Nginx

``` bash

```

``` nginx
server {
    listen 80;
    server_name yourdomain.com www.yourdomain.com;

    # Frontend
    location / {
        root /var/www/chatbot/frontend;
        index index.html;
        try_files $uri $uri/ /index.html;
    }

    # Backend API
    location /api/ {
        proxy_pass http://127.0.0.1:8000/api/;
        proxy_http_version 1.1;
        proxy_set_header Upgrade $http_upgrade;
        proxy_set_header Connection 'upgrade';
        proxy_set_header Host $host;
        proxy_cache_bypass $http_upgrade;
    }
}
```
